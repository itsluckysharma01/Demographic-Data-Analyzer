## What is the percentage of people who have a Bachelor's degree?

```python
# Calculate percentage of people with Bachelor's degree
bachelors_count = len(df[df['education'] == ' Bachelors'])
total_count = len(df)
percentage_bachelors = round((bachelors_count / total_count) * 100, 1)
percentage_bachelors
```

## What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?

```python
# Check unique education values
df['education'].unique()
```

```python
# Filter for advanced education
advanced_education = df[df['education'].isin([' Bachelors', ' Masters', ' Doctorate'])]
advanced_high_salary = advanced_education[advanced_education['salary'] == ' >50K']
percentage_advanced_high = round((len(advanced_high_salary) / len(advanced_education)) * 100, 1)
percentage_advanced_high
```

## What percentage of people without advanced education make more than 50K?

```python
# Filter for non-advanced education
non_advanced_education = df[~df['education'].isin([' Bachelors', ' Masters', ' Doctorate'])]
non_advanced_high_salary = non_advanced_education[non_advanced_education['salary'] == ' >50K']
percentage_non_advanced_high = round((len(non_advanced_high_salary) / len(non_advanced_education)) * 100, 1)
percentage_non_advanced_high
```

## What is the minimum number of hours a person works per week?

```python
min_hours = df['hours-per-week'].min()
min_hours
```

## What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?

```python
min_workers = df[df['hours-per-week'] == min_hours]
min_workers_high_salary = min_workers[min_workers['salary'] == ' >50K']
percentage_min_hours_high = round((len(min_workers_high_salary) / len(min_workers)) * 100, 1)
percentage_min_hours_high
```

## What country has the highest percentage of people that earn >50K and what is that percentage?

```python
# Calculate percentage of high earners by country
country_salary = df.groupby('native-country')['salary'].apply(lambda x: (x == ' >50K').sum() / len(x) * 100)
highest_earning_country = country_salary.idxmax()
highest_earning_percentage = round(country_salary.max(), 1)
print(f"Country: {highest_earning_country}")
print(f"Percentage: {highest_earning_percentage}%")
```

## Identify the most popular occupation for those who earn >50K in India

```python
# Filter for India and high earners
india_high_earners = df[(df['native-country'] == ' India') & (df['salary'] == ' >50K')]
most_popular_occupation = india_high_earners['occupation'].value_counts().idxmax()
most_popular_occupation
```
